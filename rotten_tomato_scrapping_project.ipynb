```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
base_site='https://editorial.rottentomatoes.com/guide/best-movies-2022/'
responce=requests.get(base_site)
responce
```

```python
html=responce.content
html
```

```python
soup=bs(html,'lxml')
with open('rottentomato.html','wb') as file:
     file.write(soup.prettify('utf-8'))
```

```python
headings=soup.find_all('div',{'class':"col-sm-18 col-full-xs countdown-item-content"})

headings
```

```python
heading=[div.find('h2')for div in headings]
heading
```

```python
heading[0]
```

```python
#extracting title,year and rating separately
movie_names=[heading.find('a').string for heading in headings]
movie_names
```

```python
#for year
years=[heading.find('span',class_='subtle start-year').string for heading in headings]
years
```

```python
#for removing the barckets in year
#use strip method
year_strip=[year.strip('()') for year in years]
year_strip
```

```python
score=[heading.find('span',class_='tMeterScore').string for heading in headings]
score
```

```python
#for directors
directors= [div.find('div',{'class':'info director'}) for div in headings]
directors
```

```python
dir=[director.find('a') for director in directors]
dir
```

```python
#to remove none valeue in director
final_directors=[None if director.find('a') is None else director.find('a').string for director in directors]
final_directors
```

```python
#for cast info
cast_info=[div.find('div',class_='info cast') for div in headings]
cast_info
```

```python
cast_info[0]
```

```python
#so we will merge each movie's cast members name in one string
cast_links=cast_info[0].find_all('a')
cast_links
```

```python
cast_names=[link.string for link in cast_links]
cast_names
```

```python
cast=",".join(cast_names)
cast
```

```python
#using for loop we merge all movies cast in one string
cast=[]
for c in cast_info:
  cast_links=c.find_all('a')
  cast_names=[link.string for link in cast_links]
  result=",".join(cast_names)
  cast.append(result)
cast
```

```python
#representing  data in structured form
import pandas as pd
data={'movie_names':[movie_names],
      'years':[year_strip],
      'score':[score],
      'cast':[cast],
      'directors':[final_directors]}
df=pd.DataFrame(data)
df
```
